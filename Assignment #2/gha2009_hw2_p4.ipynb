# Problem 4 - Linear Separability

## 4.1

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Data
X = np.array([[-1, -1], [1, 1], [-3, -3], [4, 4], [-1, 1], [1, -1], [-5, 2], [4, -8]])
y = np.array([1, 1, 1, 1, 0, 0, 0, 0])

# Plot
plt.scatter(X[:4, 0], X[:4, 1], label='Class 1')
plt.scatter(X[4:, 0], X[4:, 1], label='Class 2')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Q4.1 Dataset Visualization')
plt.show()

# Train a linear SVM
clf = SVC(kernel='linear')
clf.fit(X, y)

# Check if it classifies correctly
correctly_classified = np.all(clf.predict(X) == y)

print(f"Data is linearly separable: {correctly_classified}")
```

As seen from the scatter plot abive, the data set is not linearly separable because there is no linear function that can separate Class 1 and Class 2

One way to define $z$ is $z = x_1x_2$. If points from different classes tend to have different signs for $x_1x_2$, this transformation would make them linearly seperable. Using this mapping, points of Class 1 (−1, −1),(1, 1),(−3, −3),(4, 4) become (1), (1), (9), (6) and points of Class 2 (−1, 1),(1, −1),(−5, 2),(4, −8) become (-1), (-1), (-10), (-32).

## 4.3

```python
#Transform
z1 = [1, 1, 9, 6]
y1 = [0, 0, 0, 0]
#Plot
plt.scatter(z1, y1, label='Class 1')
z2 = [-1, -1, -10, -32]
y2 = [0, 0, 0, 0]
plt.scatter(z2, y2, label='Class 2')
plt.scatter([0], [0], label='hyperplane')
plt.xlabel('z = x1x2')
plt.legend()
plt.show()
```

**Answer:**

As seen from the above plot, now $z = 0$ separates Class 1 and Class 2. A non-linearly separable data can be represented as linearly separable form after applying a non-linear transformation($z = x_1x_2$ in this example, resulting in data represented as 1D).

## 4.4

**Answer:**

Nonlinear transformations are essential for classification problems when the dataset is not linearly separable in its original feature space. By mapping the original data into a higher-dimensional space or transforming the features, we can often make the data linearly separable in the transformed space. For instance, data points arranged in a circular pattern of two classes (one inside the circle and one outside) won't be linearly separable in 2D. However, if we transform the data using radial basis functions, they become separable in the transformed space. Nonlinear transformations thus enable linear classifiers to model complex, non-linear decision boundaries.
