```python
from vision.ssd.mobilenetv1_ssd import create_mobilenetv1_ssd
import torch.onnx
import onnx

# Parameters
net_type = 'mb1-ssd'
model_path = 'models/mb1-ssd-Epoch-99-Loss-2.7657824850082395.pth'
label_path = 'models/open-images-model-labels.txt'

# Load class names
class_names = [name.strip() for name in open(label_path).readlines()]

# Create the network and load the model
net = create_mobilenetv1_ssd(len(class_names), is_test=True)
net.load(model_path)
net.eval()

# Define paths for saving
model_onnx_path = f"models/{net_type}.onnx"

# Create dummy input and export to ONNX
dummy_input = torch.randn(1, 3, 300, 300)
torch.onnx.export(net, dummy_input, model_onnx_path, verbose=False, output_names=['scores', 'boxes'])

# Load the ONNX model and print a message
model = onnx.load(model_onnx_path)
print(f"Model ONNX format at {model_onnx_path}.")
```

## 3.4

```python
import onnx

# Load the ONNX model
onnx_model = onnx.load(model_onnx_path)

# Check the model
onnx.checker.check_model(onnx_model)
print("ONNX model is valid.")
```

## 3.5

```python
import onnxruntime

ort_session = onnxruntime.InferenceSession(model_onnx_path)

def to_numpy(tensor):
    return tensor.detach().cpu().numpy() if tensor.requires_grad else tensor.cpu().numpy()

# compute ONNX Runtime output prediction
ort_inputs = {ort_session.get_inputs()[0].name: to_numpy(dummy_input)}
ort_outs = ort_session.run(None, ort_inputs)
```

## 3.6

```python
import numpy as np

torch_out = net(dummy_input)
torch_out_np = [to_numpy(o) for o in torch_out]
```

```python
# compare ONNX Runtime and PyTorch results
np.testing.assert_allclose(torch_out_np[0], ort_outs[0], rtol=1e-03, atol=1e-05)
np.testing.assert_allclose(torch_out_np[1], ort_outs[1], rtol=1e-03, atol=1e-05)

print("Export complete.")
```

The parameters 'rtol' and 'atol' define the relative and absolute tolerances, respectively.

- rtol=1e-03 means that the relative difference between each element in the arrays should not be greater than 0.001 (0.1%).
- atol=1e-05 specifies that the absolute difference between each element should not exceed 0.00001.
If the outputs from PyTorch and ONNX Runtime match within these tolerances, the assertions will pass, indicating that the outputs are indeed close to each other to the specified precision.

```python
from PIL import Image
import numpy as np
import onnxruntime as ort

# Load and preprocess the image
def preprocess_image(image_path):
    input_shape = (1, 3, 300, 300)
    img = Image.open(image_path)
    img = img.resize((300, 300), Image.BILINEAR)

    img_data = np.array(img)
    img_data = np.transpose(img_data, [2, 0, 1])
    img_data = np.expand_dims(img_data, 0)

    mean_vec = np.array([0.485, 0.456, 0.406])
    stddev_vec = np.array([0.229, 0.224, 0.225])
    norm_img_data = np.zeros(img_data.shape).astype('float32')
    for i in range(img_data.shape[1]):
        norm_img_data[:,i,:,:] = (img_data[:,i,:,:]/255 - mean_vec[i]) / stddev_vec[i]

    return norm_img_data, img

# Load the ONNX model using ONNX Runtime
ort_session = ort.InferenceSession("models/mb1-ssd.onnx")

# Load and preprocess images - will be shown in next part
image_paths = ['/content/pytorch-ssd/data/open_images/test/310537222f4d8c1d.jpg',
               '/content/pytorch-ssd/data/open_images/test/44342a60232e50cb.jpg']
for image_path in image_paths:
    img_y, img = preprocess_image(image_path)

    # Inference
    ort_inputs = {ort_session.get_inputs()[0].name: img_y}
    ort_outs = ort_session.run(None, ort_inputs)

    # Process the output
    print(ort_outs[0])
```

From the matrix above, we can see a mix of high and low confidence scores across different elements in each image. For example, in some cases, the model is almost certain about its classification (scores close to 1), while in others, the confidence is lower (scores distributed more evenly across classes).

```python
from PIL import Image
import numpy as np
import onnxruntime as ort
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Load and preprocess the image
def preprocess_image(image_path):
    input_shape = (1, 3, 300, 300)
    img = Image.open(image_path)
    img = img.resize(input_shape[2:], Image.BILINEAR)

    img_data = np.array(img)
    img_data = np.transpose(img_data, [2, 0, 1])
    img_data = np.expand_dims(img_data, 0)

    mean_vec = np.array([0.485, 0.456, 0.406])
    stddev_vec = np.array([0.229, 0.224, 0.225])
    norm_img_data = np.zeros(img_data.shape).astype('float32')
    for i in range(img_data.shape[1]):
        norm_img_data[:,i,:,:] = (img_data[:,i,:,:]/255 - mean_vec[i]) / stddev_vec[i]

    return norm_img_data, img

# Define your class names
class_names = ["Handgun", "Shotgun"]  # Update as per your classes

# Load the ONNX model using ONNX Runtime
ort_session = ort.InferenceSession("models/mb1-ssd.onnx")

# Load, preprocess, and annotate images
image_paths = ['/content/pytorch-ssd/data/open_images/test/44342a60232e50cb.jpg']
#
#/content/pytorch-ssd/data/open_images/test/49e40ccefbef1c86.jpg

for image_path in image_paths:
    norm_img_data, original_img = preprocess_image(image_path)

    # Inference
    ort_inputs = {ort_session.get_inputs()[0].name: norm_img_data}
    ort_outs = ort_session.run(None, ort_inputs)

    # Extract bounding boxes, labels, scores
    labels_scores = ort_outs[0][0]   # Labels and scores
    bboxes = ort_outs[1][0]          # Bounding boxes

    # Assuming the first index is label and the second index is score
    labels = labels_scores[:, 0]
    scores = labels_scores[:, 1]

    # Display image with bounding boxes and labels
    # Assuming the original image size is the one used for detection
    original_img_width, original_img_height = original_img.size

    plt.figure(figsize=(8, 8))
    plt.imshow(original_img)
    ax = plt.gca()

    # Loop through the detections and plot the bounding boxes
    count = 0
    for box, label, score in zip(bboxes, labels, scores):

        if score > 0.05 and count < 3 and round(label)==1:  # Threshold can be adjusted
            count+=1
            #print(f"Label: {round(label)}, Score: {score:.2f}")
            # Scale the bounding box back to the original image size
            xmin, ymin, xmax, ymax = box
            #print(box)
            xmin *= original_img_width
            xmax *= original_img_width
            ymin *= original_img_height
            ymax *= original_img_height

            # Calculate width and height of the box
            box_width = xmax - xmin
            box_height = ymax - ymin

            # Create the rectangle for the bounding box
            bbox = patches.Rectangle((xmin, ymin), box_width, box_height, linewidth=2, edgecolor='red', facecolor='none')
            ax.add_patch(bbox)

            # Add a label with the class name and score
            label_text = f"{class_names[round(label)]}"
            plt.text(xmin, ymin, label_text, color='white', verticalalignment='top',
                    bbox={'color': 'red', 'pad': 0})


    plt.axis('off')
    plt.show()
```
